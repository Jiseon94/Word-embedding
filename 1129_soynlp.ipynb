## 품사판별기 POS 의 아쉬운점

```python
okt = Okt()
okt.pos('아이오아이가 5년뒤에 다시 모였으면 좋겠습니다. 소나기 곡 너무 찡해요')
```

```python
hannanum.pos('아이오아이가 5년뒤에 다시 모였으면 좋겠습니다. 소나기 곡 너무 찡해요')
```

품사 판별기는 주로 말뭉치라 불리는 학습데이터를 기반으로 문장/어절의 패턴을 학습.
학습용 데이터를 이용하는 방법을 supervised learning이라 합니다. 하지만, supervised learning 기반 품사 판별기가 가지는 위험 중 하나는 모르는 단어가 등장했을 때 이를 처리하는 방법입니다. '아이오아이' 분명 단어임에도 학습 때 본 적이 없다면 단어로 인식되지 않을 수 있습니다.

만약 영화리뷰를 분석하고 있다면 '끝까지간다'라는 단어가 여러 번 등장할 것이고, 우리는 리뷰들에 있는 이 단어에 여러번 노출되어 단어로 인식할 것입니다. 하지만 품사 판별기의 목적은 주어진 문장/어절을 알고 있는 단어들로 분해하여 인식하는 것이기 때문에 이를 하나의 단어로 인식하지 않습니다.

```python
okt.pos('끝까지간다')
```

```python
hannanum.pos('끝까지간다')
```

그래서 주어진 문서 집합에서 패턴을 찾아내어 단어를 추출해보려 합니다. 다른 학습데이터는 이용하지 않으며, 통계에 기반하여 단어를 찾아내는 unsupervised learning 방법입니다.

- Accessor Variety
- Branching Entropy
- Cohesion Score

## Branching Entropy

주어진 단어의 불확실성이 내려가면 -> entropy 내려감
- ex) n_  ? 불확실성 큼 / natur_ ? : e 가 오거나, al 이 올 수 있으니 불확실성이 내려감

근데, nature 이라는 단어가 오고, 그 다음을 추측하라고 하면, 무수히 많은 단어들이 올 수 있기에 다시 불확실성이 올라감. 단어의 경계에서는 다시 entropy 가 올라가는 현상.

## Accessor Variety

branching entropy의 discrete 버전.

nature, natural 두 단어가 존재한다고 하면 natur + [-e, -a]이기 때문에 right-side의 accessor variety는 len({-e, -a}) = 2 입니다. Branching entropy는 -e와 -a가 몇 번 나왔는지의 빈도수를 센 다음, 이를 이용하여 entropy를 계산하는게 차이입니다.

## Cohesion Score

##### 한국어의 어절은 L + [R] 구조입니다. 띄어쓰기가 제대로 되어있다면 한국어는 의미를 지니는 단어 (명사,동사,형용사,부사)가 어절의 왼쪽 (L part)에, 문법기능을 하는 (조사,어미)가 어절의 오른쪽에 등장합니다.

단어를 구성하는 글자들이 얼마나 함께 자주 등장하는지의 정보로도 단어의 경계를 판단할 수 있습니다.

```python
import sys
sys.path.insert(0, '../')
from soynlp.utils import DoublespaceLineCorpus
```

```python
corpus_fname = '2016-10-20.txt'
```

```python
sentences = DoublespaceLineCorpus(corpus_fname, iter_sent=True)
print('num sentences = %d' % len(sentences))
```

```python
from soynlp.word import WordExtractor

word_extractor = WordExtractor(
    min_frequency=100,
    min_cohesion_forward=0.05, 
    min_right_branching_entropy=0.0
)

word_extractor.train(sentences)
words = word_extractor.extract()
```

```python
len(words)
# words는 {word:Score} 형식의 dictionary
```

```python
print('type: %s\n' % type(words['아이오아이']))
print(words['아이오아이'])
```

```python
import math

def word_score(score):
    return (score.cohesion_forward * math.exp(score.right_branching_entropy))

print('단어   (빈도수, cohesion, branching entropy)\n')
for word, score in sorted(words.items(), key=lambda x:word_score(x[1]), reverse=True)[:30]:
    print('%s     (%d, %.3f, %.3f)' % (
            word, 
            score.leftside_frequency, 
            score.cohesion_forward,
            score.right_branching_entropy
            )
         )
```

```python
cohesion_scores = word_extractor.all_cohesion_scores()
cohesion_scores['아이오아이'] # (cohesion_forward, cohesion_backward)
```

```python
branching_entropy = word_extractor.all_branching_entropy()
branching_entropy['아이오아이'] # (left_branching_entropy, right_branching_entropy)
```

```python
accessor_variety = word_extractor.all_accessor_variety()
accessor_variety['아이오아이'] # (left_accessor_variety, right_accessor_variety)
```
